### "F test of variance ratio"

```python
import numpy as np
import scipy.stats as st
```

#### Answer the following questions by providing Python code:

In a factory there are two packaging machines. Output samples are drawn from each machine. <br>
- $n_1=15$, $\bar{x}_1=5.0592 kg$, $s_1^2=0.1130 kg^2$
- $n_2=12$, $\bar{x}_2=4.9808 kg$, $s_2^2=0.0152 kg^2$  

Test whether there is a significant difference in the variances (95%).

```python
n1 = 15
ssq1 = 0.1130
n2 = 12
ssq2 = 0.0152
```

1). Do the right tail test:

```python
test_stat = ssq1/ssq2
1 - st.f.cdf(test_stat, n1-1, n2-1)
```

2). Do the two tail test:

```python
test_stat = ssq1/ssq2
(1 - st.f.cdf(test_stat, n1-1, n2-1))*2
```

3). Briefly discuss the results:

A **right tailed test** (sometimes called an upper test) is where your hypothesis statement contains a greater than (>) symbol. In other words, the inequality points to the right.

A **two tailed test** tells you that youâ€™re finding the area in the middle of a distribution. In other words, your rejection region (the place where you would reject the null hypothesis) is in both tails.

In right tailed test, the entire 5% would be in a single tail. But with a two tailed test, that 5% is split between the two tails, giving you 2.5% (0.025) in each tail.

To interpret the results, we compare the p-value to our significance level.

For the right tailed test we got a p-value = 0.0009847696603404454 and it's less than the significance level.

Same thing for the two tailed test , we got p-value = 0.001969539320680891 and it's less than the significance level.
