### "Policy support rate"

```python
import numpy as np
import scipy.stats as st
```

#### Answer the following questions by providing Python code:

It is known that the overall policy support rate is 52%. <br>

```python
p = 0.52
```

1). What is the probability of support rate less than 50% when asked to 100 people? <br>
HINT: Use sample proportion and Standard Normal distribution.

```python
n1 = 100
SE =np.sqrt(p*(1-p)/n1)
Z = (0.5-p)/SE
st.norm.cdf(Z)
```

2). What is the probability of support rate less than 50% when asked to 300 people? <br>
HINT: Use sample proportion and Standard Normal distribution.

```python
n2 = 300
SE =np.sqrt(p*(1-p)/n2)
Z = (0.5-p)/SE
st.norm.cdf(Z)
```

3). Redo the calculations of 1) and 2) using the Binomial distribution. Then, compare.

```python
# Redo of 1).
st.binom.cdf(0.5*n1,n1,p)
```

```python
# Redo of 2).
st.binom.cdf(0.5*n2,n2,p)
```

The results we get using Binomial distribution are a litlle bit bigger than the results we got using sample proportion and Standard Normal distribution.

4). Redo the calculations of 1) and 2) using the Poisson distribution. Then, compare.

```python
# Redo of 1).
位 = n1 * p
r = 0.5 * n1
st.poisson.cdf(r,位)
```

```python
# Redo of 2).
位 = n2 * p
r = 0.5 * n2
st.poisson.cdf(r,位)
```

The results we get using Poisson distribution are a litlle bit bigger than the results we got using  Binomial distribution and Standard Normal distribution.
